## Understanding Random Forest Algorithm
Random Forest is an ensemble learning technique used for classification and regression tasks. It creates multiple decision trees during training and merges their outputs to improve accuracy and control overfitting.

**Key Features of Random Forest:**
- It is based on bagging (Bootstrap Aggregating), where multiple decision trees are trained on random subsets of data.
- The final prediction is determined through voting (for classification) or averaging (for regression).
- It helps reduce overfitting compared to a single decision tree.

**Why Random Forest?**
- Handles missing values efficiently.
- Works well with large datasets and high-dimensional data.
- Provides feature importance, which helps in understanding which variables influence predictions the most.

# Buliding Model

```python
X=df.drop('diagnosis', axis=1)
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train ,y_test =train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
# scaling data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

rand_clf = RandomForestClassifier(criterion = 'entropy', max_depth = 10, max_features = 0.5, min_samples_leaf = 2, min_samples_split = 3, n_estimators = 130)
rand_clf.fit(X_train, y_train)
```

```python
# from sklearn.ensemble import GradientBoostingClassifier
# rand_clf = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=5)
# rand_clf.fit(X_train, y_train)
```

```python
y_pred = rand_clf.predict(X_test)
```

```python
# Save the trained model to a file
joblib_file = "random_forest_model.pkl"
joblib.dump(rand_clf, joblib_file)

# Print confirmation message
print(f"Model saved to {joblib_file}")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(accuracy_score(y_train, rand_clf.predict(X_train)))
rand_clf_acc = accuracy_score(y_test, rand_clf.predict(X_test))
print(rand_clf_acc)
y_pred = rand_clf.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### Visualize a Single Decision Tree
Below is a visualization of one of the decision trees in the Random Forest. It shows how the tree splits the data based on different features to classify the tumors as benign (0) or malignant (1). For clarity, we limit the tree depth to 3 to avoid overcrowding the plot.

```python
from sklearn import tree
import matplotlib.pyplot as plt

# Get the first tree from the Random Forest
single_tree = rand_clf.estimators_[0]

plt.figure(figsize=(20,10))
tree.plot_tree(single_tree, 
               feature_names=X.columns, 
               class_names=['Benign', 'Malignant'], 
               filled=True, 
               rounded=True, 
               max_depth=3)  # Limit to depth 3 for readability
plt.title("Visualization of a Single Decision Tree (Limited to Depth 3)")
plt.show()
```

### Feature Importances
The following bar plot displays the importance of each feature in the classification process, as determined by the Random Forest algorithm. Feature importance is based on how much each feature contributes to reducing impurity across all trees.

```python
import numpy as np

importances = rand_clf.feature_importances_
indices = np.argsort(importances)[::-1]  # Sort in descending order

plt.figure(figsize=(12,6))
plt.title("Feature Importances")
plt.bar(range(X.shape[1]), importances[indices], align="center")
plt.xticks(range(X.shape[1]), X.columns[indices], rotation=90)
plt.xlim([-1, X.shape[1]])
plt.xlabel("Features")
plt.ylabel("Importance")
plt.tight_layout()
plt.show()
```

### Voting Process for Test Samples
Below, we demonstrate how individual trees in the Random Forest vote for the class (Benign or Malignant) for the first 5 test samples. For each sample, we show the number of trees voting for each class and the final prediction based on majority voting. Bar charts visualize the voting distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Select the first 5 test samples
X_test_samples = X_test[:5]
y_test_samples = y_test[:5]

# Get predictions from each tree
tree_predictions = np.array([tree.predict(X_test_samples) for tree in rand_clf.estimators_])

# Transpose to get predictions per sample
tree_predictions = tree_predictions.T

# For each sample, print predictions and majority vote
for i, (sample_preds, true_label) in enumerate(zip(tree_predictions, y_test_samples)):
    print(f"Sample {i}:")
    print(f"True label: {'Malignant' if true_label == 1 else 'Benign'}")
    print(f"Tree predictions: {sample_preds}")
    malignant_votes = np.sum(sample_preds == 1)
    benign_votes = np.sum(sample_preds == 0)
    final_prediction = 1 if malignant_votes > benign_votes else 0
    print(f"Malignant votes: {malignant_votes}, Benign votes: {benign_votes}")
    print(f"Final prediction: {'Malignant' if final_prediction == 1 else 'Benign'}")
    print()

# Visualize voting distribution for each sample
for i, sample_preds in enumerate(tree_predictions):
    malignant_votes = np.sum(sample_preds == 1)
    benign_votes = np.sum(sample_preds == 0)
    true_label = y_test_samples.iloc[i]  # Use .iloc[i] to access by position
    plt.figure(figsize=(6,4))
    plt.bar(['Benign', 'Malignant'], [benign_votes, malignant_votes], color=['blue', 'red'])
    plt.title(f"Voting for Sample {i} (True Label: {'Malignant' if true_label == 1 else 'Benign'})")
    plt.ylabel('Number of Trees')
    plt.show()
```

### Accuracy of Individual Trees vs. Ensemble
The histogram below shows the distribution of accuracies of individual trees on the test set, with the ensemble's accuracy marked by a red dashed line. This visualization demonstrates the power of ensemble learning, as the ensemble typically outperforms most individual trees.

```python
from sklearn.metrics import accuracy_score

# Calculate accuracy of each individual tree
individual_tree_accuracies = []
for tree in rand_clf.estimators_:
    tree_pred = tree.predict(X_test)
    tree_acc = accuracy_score(y_test, tree_pred)
    individual_tree_accuracies.append(tree_acc)

# Ensemble accuracy (already calculated)
ensemble_acc = accuracy_score(y_test, y_pred)

plt.figure(figsize=(10,6))
plt.hist(individual_tree_accuracies, bins=20, alpha=0.7, label='Individual Trees', color='green')
plt.axvline(ensemble_acc, color='red', linestyle='dashed', linewidth=2, label=f'Ensemble (Accuracy = {ensemble_acc:.2f})')
plt.xlabel('Accuracy')
plt.ylabel('Number of Trees')
plt.title('Accuracy of Individual Trees vs. Ensemble')
plt.legend()
plt.show()
```

### Confusion Matrix
The heatmap below visualizes the confusion matrix, showing the number of true positives (TP), true negatives (TN), false positives (FP), and false negatives (FN). This helps assess the model's performance.

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Benign', 'Malignant'], 
            yticklabels=['Benign', 'Malignant'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
